```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import pathlib
```

```python
LogData = '/Users/johnrandelin/Documents/Taigatech/Utveckling/HilmerData/Meassure/August_28_Between_15_And_16/LogData.csv'
Ej_Korregerat = '/Users/johnrandelin/Documents/Taigatech/Utveckling/HilmerData/Meassure/August_28_Between_15_And_16/EJ_Korregerat'
LogData = pd.read_csv(LogData)
#This is a test comment
#This is a second test comment
```

```python
#Collecting the filepaths
filepaths = []
for index, row in LogData.iterrows(): 
    ID = row['LogID']
    path = Ej_Korregerat+'/'+ID+'.csv'
    filepaths.append(path)
```

```python
#Now we are ready to plot
NumGraphs = 3 #How many graphs do you need to plot? 

for i in range(0,NumGraphs): 
    Log = pd.read_csv(filepaths[i])
    plt.plot(range(0,len(Log)), Log['Stereo Vinklar'], label = 'Stereo')
    plt.plot(range(0,len(Log)), Log['PLC Vinklar'], label = 'PLC')
    path = pathlib.PurePath(filepaths[i])
    plt.title(path.name)
    plt.legend()
    plt.grid()
    plt.show()
```
